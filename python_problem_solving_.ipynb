```python
class Account:
    def __init__(self,owner,balance=0):
        self.owner = owner
        self.balance = balance

    def __str__(self):
        return f'Account owner:   {self.owner}\nAccount balance: ${self.balance}'

    def deposit(self,dep_amt):
        self.balance += dep_amt
        print('Deposit Accepted')

    def withdraw(self,wd_amt):
        if self.balance >= wd_amt:
            self.balance -= wd_amt
            print('Withdrawal Accepted')
        else:
            print('Funds Unavailable!')

acct1 = Account('Jose',100)
print(acct1)

acct1.owner

acct1.balance

```

```python
class Line(object):

    def __init__(self,coor1,coor2):
        self.coor1 = coor1
        self.coor2 = coor2

    def distance(self):
        x1,y1 = self.coor1
        x2,y2 = self.coor2
        return ((x2-x1)**2 + (y2-y1)**2)**0.5

    def slope(self):
        x1,y1 = self.coor1
        x2,y2 = self.coor2
        return (y2-y1)/(x2-x1)
```

```python
coordinate1 = (3,2)
coordinate2 = (8,10)

li = Line(coordinate1,coordinate2)

li.distance()
```

```python
class PaymentError(Exception):
    """Custom exception for payment errors."""
    pass

def process_payment(amount):
    if amount <= 0:
        raise ValueError("Error: Payment amount must be greater than zero.")
    if amount > 5000:  # Simulating insufficient balance
        raise PaymentError("Error: Insufficient balance for this transaction.")

    # Simulating a successful payment
    print(f"Payment of {amount} processed successfully!")

try:
    amount = float(input("Enter the payment amount: "))
    process_payment(amount)
except ValueError as ve:
    print(ve)
except PaymentError as pe:
    print(pe)
except Exception as e:
    print(f"Unexpected error: {e}")
finally:
    print("Thank you for using our payment system.")
```

```python
# using getter and setter
class Person:
  def __init__(self,name,age):
    print("constructor")
    self.__name = name
    self.__age = age
  def show(self):
    print("my name is ",self.name," and my age is ",self.age)

  @property
  def name(self):
    return self.__name

  @name.setter
  def name(self,value):
    self.__name = value
    return self.__name
  @property
  def age(self):
    return self.__age

  @age.setter
  def age(self,value):
    self.__age = value
    return self.__age


p = Person("xyz",29)
print(p.name)
print(p.age)
p.show()
p.name = "aman"
p.age = 30
print(p.name)
print(p.age)
p.show()
```

```python
class Fraction:
  def __init__(self,n,d):
    self.num = n
    self.den = d
  def __str__(self):
    return "{}/{}".format(self.num,self.den)
  def __add__(self,other):
    new_num = self.num * other.den + self.den * other.num
    new_den = self.den * other.den
    return Fraction(new_num,new_den)
  def __sub__(self,other):
    new_num =self.num*other.den - self.den * other.num
    new_den = self.den*other.den
    return Fraction(new_num,new_den)
x = Fraction(2,3)
y = Fraction(4,5)
print(x+y)
print(x-y)
```

```python
class Vehicle:  # Base class
    def __init__(self, make, model):
        print("vehicle class constructor")
        self.make = make
        self.model = model

    def start(self):
        print("Engine started")

class Car(Vehicle):  # Intermediate class inheriting from Vehicle
    def __init__(self, make, model, num_doors):
        print("car class constructor")
        super().__init__(make, model)
        self.num_doors = num_doors

    def drive(self):
        print("Driving...")

class ElectricCar(Car):  # Derived class inheriting from Car
    def __init__(self, make, model, num_doors, battery_capacity):
        print("electric class constructor")
        super().__init__(make, model, num_doors)
        self.battery_capacity = battery_capacity

    def charge(self):
        print("Charging...")

my_electric_car = ElectricCar("Tesla", "Model S", 4, "100 kWh")

print(my_electric_car.make)       # Access from Vehicle
print(my_electric_car.num_doors)  # Access from Car
print(my_electric_car.battery_capacity)  # Access from ElectricCar

my_electric_car.start()  # From Vehicle
my_electric_car.drive()   # From Car
my_electric_car.charge()  # From ElectricCar
```

```python
from abc import abstractmethod
class Vehicle:  # Base class (still the same)
    def __init__(self, make, model):
        print("vehicle constuctor called")
        self.make = make
        self.model = model
    @abstractmethod
    def start(self):
        pass

class Car(Vehicle):  # Intermediate class (still the same)
    def __init__(self, make, model, num_doors):
        print(" car constuctor called")
        super().__init__(make, model)
        self.num_doors = num_doors
    def start(self):
        print("engine started")
    def drive(self):
        print("Driving...")

class Electric:  # New independent class for electric functionality
    def __init__(self, battery_capacity):
        print("electric  constuctor called")
        self.battery_capacity = battery_capacity

    def charge(self):
        print("Charging...")

class ElectricCar(Car, Electric):  # Multiple inheritance!
    def __init__(self, make, model, num_doors, battery_capacity):
        print("electric car constuctor called")
        Car.__init__(self, make, model, num_doors) #Explicitly calling car init.
        Electric.__init__(self, battery_capacity) #Explicitly calling Electric init.
        # super().__init__(make, model, num_doors)  #Not required.

    # No need to redefine charge() as it's inherited from Electric

my_electric_car = ElectricCar("Tesla", "Model S", 4, "100 kWh")

print(my_electric_car.make)       # From Vehicle
print(my_electric_car.num_doors)  # From Car
print(my_electric_car.battery_capacity)  # From Electric

my_electric_car.start()  # From Vehicle
my_electric_car.drive()   # From Car
my_electric_car.charge()  # From Electric

print(ElectricCar.__mro__) #Prints method resolution order.
```

```python
# sum of  min and max
def min_max(a):
  n = len(a)
  if n == 1 :
    return (a[0],a[0])
  if a[0] > a[1]:
    maximum = a[0]
    minimum = a[1]
  else:
    maximum = a[1]
    minimum = a[0]

  for i in range(2,n):
    if a[i]>maximum:
      maximum = a[i]
    elif a[i]<minimum:
      minimum = a[i]
  return (maximum+minimum)
arr = [-2]
y = min_max(arr)
print(y)
```

```python
def arrange(a):

  n = len(a)
  low = 0
  high = n-1

  while low <= high:
    if a[low] <= 0:
      low += 1
    elif a[high] >= 0:
      high -= 1
    else:
      a[low],a[high] =a[high],a[low]
      low += 1
      high -= 1
  return(a)

x = [ 6,-5,-7,-9,3,5,-8,1,9,-3]
arrange(x)
```
